### 2. Which sport is having most number of Gold Medals so far? (Top 5)

---



```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("summer.csv")
```

```python
df.sort_values("Sport", inplace = True) 
mul_filter1 = df["Medal"]=="Gold"
df.where(mul_filter1, inplace = True) 
df.dropna()
```

```python
A = df['Sport'].value_counts().head()
print(A)
```

```python
x = A.keys()
x = x.array
y = A.values

plt.bar(x, y)
 
plt.xlabel("Sports")
plt.ylabel("No. of Gold medals")
plt.title("Number of Gold Medals earned by the Top 5 sports")
plt.show()

print('The Sport having the most number of Gold Medals is "AQUATICS"')
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("summer.csv")
```

```python
df['Sport'].unique()
var = df.groupby(['Sport'], sort=False)['Medal'].count()
sport_max = var.nlargest(5)
print(sport_max)
```

```python
x = sport_max.keys()
x = x.array
y = sport_max.values

plt.bar(x, y)
 
plt.xlabel("Sports")
plt.ylabel("No. of medals")
plt.title("Number of Medals earned by the Top 5 sports")
plt.show()
print('The Sport having the most number of Medals is "AQUATICS"')
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("summer.csv")
```

```python
mul_filter1 = df["Medal"]=="Gold"
mul_filter2 = df["Country"] == "IND"
df.where(mul_filter1 & mul_filter2, inplace = True) 
print(df.dropna())
print("\n")
print("India won the first Gold medal in summer olympics in the Year 1928.")
```

### 7. Which event is most popular in terms on number of players?

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("summer.csv")
```

```python
data = []

for event in df['Event'].unique():
    data.append([event , len(df[df['Event']  == event])])

pd.DataFrame(data,columns = ['Event','freq']).sort_values(by='freq', ascending=False).plot(x = 'Event', y = 'freq', kind = 'bar', figsize = (100,100))
print("\n")
print("The most popular event in terms of number of players is FOOTBALL.")
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("summer.csv")
```

```python
mul_filter1 = df["Medal"]=="Gold"
mul_filter2 = df["Gender"]=="Women"
df.where(mul_filter1 & mul_filter2, inplace = True) 
print(df.dropna())
```

```python
female = df['Sport'].value_counts()
female_players_medals = female.nlargest(5)
print(female_players_medals)
```

```python
x = female_players_medals.keys()
x = x.array
y = female_players_medals.values

plt.bar(x, y)
 
plt.xlabel("Female Players")
plt.ylabel("No. of Gold medals")
plt.title("Number of GoldMedals earned by the Top 5 Female Players")
plt.show()
print('The Sport having the most number of Gold Medals won by Females is "AQUATICS"')
```
